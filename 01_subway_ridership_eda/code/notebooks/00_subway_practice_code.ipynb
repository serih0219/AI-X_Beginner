```python
# 월별 승하차총승객수 계산
# 가설) 1월~6월 중에 5월에 지하철 승객수가 많다? -맞음!
# 이유?? 
monthly_passengers = raw.groupby('연월')['승하차총승객수'].sum().reset_index()

print(monthly_passengers)

max_monthly_passengers=monthly_passengers.idxmax()

print(max_monthly_passengers)


plt.figure(figsize=(12, 8))
sns.barplot(data=monthly_passengers, x='연월', y='승하차총승객수')

plt.title(' < 2019년 월별 승하차총승객수 > ')
plt.xlabel(' < 년도 / 월 > ')
plt.ylabel(' < 승 하 차 총 승 객 수 >')
plt.xticks(fontsize=12)

plt.show()
```

```python

# 요일별 승하차총승객수 합계 계산
daily_passengers = raw.groupby('요일')['승하차총승객수'].sum().reset_index()

# 요일 순서를 '월화수목금토일'로 정렬
day_order = ['월', '화', '수', '목', '금', '토', '일']
daily_passengers['요일'] = pd.Categorical(daily_passengers['요일'], categories=day_order, ordered=True)

# 최대 승객수 요일 찾기
max_daily_passengers = daily_passengers.loc[daily_passengers['승하차총승객수'].idxmax()]
print(max_daily_passengers)

# 그래프 그리기
plt.figure(figsize=(12, 8))
sns.barplot(data=daily_passengers, x='요일', y='승하차총승객수', order=day_order)
plt.title(' < 요일별 승하차총승객수 > ')
plt.xlabel(' < 요 일 > ')
plt.ylabel(' < 승하차총승객수 > ')
plt.xticks(fontsize=12)
plt.show()
```

```python
# Q) 가장 승객이 많이 타는 승차역은? (승차총승객수 계산!) 답:잠실

amount_of_people_station=raw.groupby('역명')['승차총승객수'].sum()
print(amount_of_people_station)

most_popular_station=amount_of_people_station.idxmax()
most_passengers = amount_of_people_station.max()


print(f"승차역: {most_popular_station}, 승객 수: {most_passengers}")
```

```python
##Q) 1호선에서 가장 하차를 많이 하는 역은? (1호선에서는 가산디지털단지)


line_1_data = raw[raw['노선명'] == '1호선']

line_1_data_station=line_1_data.groupby('역명')['하차총승객수'].sum()

line_1_data_maxout_station=line_1_data_station.idxmax()
line_1_data_maxout_passengers=line_1_data_station.max()

print(f"하차역:{line_1_data_maxout_station}, 하차승객 수:{line_1_data_maxout_passengers}")

```

```python

##노선별로 역별/요일별 승차승객수를 비교해보자
##1호선 역별 그래프
##맥북은 애플고딕..
plt.rcParams['font.family'] = 'AppleGothic'
##1호선 그래프를 그리기 때문에 노선명이 1호선인 데이터만 뽑고 역별/요일별 그래프이기 때문에 
##groupby를 사용해서 역명,요일별로 승차총승객수만 모두 더했고, reset_index()를 사용해서 인덱스 초기화하기 (그래야 '승차총승객수'가 인덱스가 아닌 열로 존재!)
line1_data=raw[raw['노선명']=='1호선']
line1_data_station=line1_data.groupby(['역명','요일'])['승차총승객수'].sum().reset_index()
##heatmap에 사용하기 위해서 pivot table 생성해주기
line1_data_station=pd.pivot_table(line1_data_station,index='역명',columns='요일',values='승차총승객수',fill_value=0)
##요일 순서대로 명시해주기
line1_data_station = line1_data_station[['월', '화', '수', '목', '금', '토', '일']]

plt.subplots(figsize=(7,12))
sns.heatmap(line1_data_station,cmap='YlGnBu',linewidth=0.5)
plt.yticks(ticks=np.arange(len(line1_data_station.index)) + 0.5, labels=line1_data_station.index)
plt.tight_layout()
plt.title('< 1 호 선 >')
plt.xlabel('< 요 일 >')
plt.ylabel('< 역 이 름 >')
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
# '사용일자'를 datetime 형식으로 변환
raw['사용일자'] = pd.to_datetime(raw['사용일자'])

# 1월부터 6월 데이터만 필터링
raw_filtered = raw[(raw['사용일자'].dt.month >= 1) & (raw['사용일자'].dt.month <= 6)]

# '역명'을 기준으로 데이터 병합
merged = pd.merge(raw_filtered, raw1, left_on='역명', right_on='역이름')

# 승하차총승객수를 역별로 합산
total_passengers = merged.groupby(['역명', '위도', '경도']).agg({'승하차총승객수': 'sum'}).reset_index()

# 총 승객 수 기준 상위 10개 역 추출
top_10_stations = total_passengers.nlargest(10, '승하차총승객수')

# 지도 생성 (서울 중심으로 초기화)
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# 히트맵을 위한 데이터 준비 (위도, 경도, 가중치)
heat_data = [
    [row['위도'], row['경도'], row['승하차총승객수']]
    for index, row in total_passengers.iterrows()  # 모든 역의 데이터 사용
]

# 히트맵 추가 (승객수를 가중치로 사용)
HeatMap(heat_data, 
        max_zoom=15, 
        radius=25,       # 반경을 늘려서 가시성 증가
        blur=15,         # 블러 효과 적용
        min_opacity=0.5  # 최소 불투명도 설정
       ).add_to(m)

# 중복되지 않은 상위 10개 역에 마커 및 이름 추가
for index, row in top_10_stations.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['역명']} (총 승객수: {int(row['승하차총승객수'])})",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)

# 지도 출력
m.save('subway_heatmap_top10.html')
m
```

```python
import pandas as pd
import folium

# 승차총승객수 기준 상위 20개의 역을 찾기 위해 raw 데이터 정렬
top_20_stations = raw.groupby('역명')['승차총승객수'].sum().nlargest(20).reset_index()

# 순위 컬럼 추가
top_20_stations['순위'] = top_20_stations['승차총승객수'].rank(ascending=False, method='min').astype(int)

# 역명 기준으로 raw1과 병합
merged_df = pd.merge(top_20_stations, raw1, left_on='역명', right_on='역이름')

# Folium 지도 생성 (서울 기준으로 설정)
m = folium.Map(location=[37.527, 126.9760], zoom_start=12)

# 환승역 리스트
transfer_stations = [
    '신도림', '사당', '잠실', '홍대입구', 
    '강남', '고속터미널', '선릉', '서울역','종로3가','시청','건대입구','교대',
    '가산디지털단지',
]

# 버스터미널 주변 역 리스트
bus_terminal_stations = [
    '강변', '고속터미널', '남부터미널', '서울역'
]

# 상위 20개의 역에 대한 마커 추가 (순위 포함)
for idx, row in merged_df.iterrows():
    # 마커 색상 설정
    if row['역명'] in bus_terminal_stations:
        marker_color = 'red'  # 버스터미널 주변 역 색상
    elif row['역명'] in transfer_stations:
        marker_color = 'orange'  # 환승역 색상
    else:
        marker_color = 'gray'  # 기본 색상
    
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"순위 {row['순위']}: {row['역명']} - {row['승차총승객수']}명",
        tooltip=f"{row['순위']}위: {row['역명']}",
        icon=folium.Icon(color=marker_color)  # 색상 적용
    ).add_to(m)

# 각 호선의 색상 설정
line_colors = {
    '1호선': 'navy',
    '2호선': 'green',
    '3호선': 'tomato',
    '4호선': 'skyblue',
    '5호선': 'purple',
    '6호선': 'brown',
    '7호선': 'khaki',
    '8호선': 'lightpink',
    '9호선': 'gray',  # 신분당선 색상 설정
    '신분당선': 'red',
    '수인선':'gold'
}

# 호선 별로 선 그리기
for line_name, line_stations in {
    '1호선': line_1_stations,
    '2호선': line_2_stations,
    '3호선': line_3_stations,
    '4호선': line_4_stations,
    '5호선': line_5_stations,
    '6호선': line_6_stations,
    '7호선': line_7_stations,
    '8호선': line_8_stations,
    '9호선': line_9_stations,
    '신분당선': 신분당선_stations,
    '수인선':suwon_line_stations # 신분당선 추가
}.items():
    line_coords = []
    for station in line_stations:
        station_data = raw1[raw1['역이름'] == station]
        if not station_data.empty:
            line_coords.append((station_data['위도'].values[0], station_data['경도'].values[0]))

    # 선 색상 설정
    if line_coords:
        folium.PolyLine(locations=line_coords, color=line_colors[line_name], weight=5, opacity=1).add_to(m)

# 지도 출력
m.save("top_20_stations_map_with_bus_terminals_final.html")
m
```
